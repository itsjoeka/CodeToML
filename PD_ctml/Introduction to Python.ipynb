### Casting
Casting can be used to specify data types. When assigning a value to a variable, casting can be used to indicate if the variable should be returned as a string, an integer or a float. Strings, intergers, and floats will be discussed further under python data types

```python
x = str(3) # will be returned as a string ('3')
y = int(3) # will be rturned as an integer (3)
z = float(3) # will be returned as a float (3.0)
print(x, y, z)
```

### Assigning Multiple Variables

#### Multiple values to Multiple variables
Python allows you to assign many vales to multiple variable at once. 

```python
a, b, c, = "Patano", "money", "play"
print(a) 
print(b)
print(c)
```

#### One value to Mutiple variables
In Python, we can assign one value to many variables.

```python
a = b = c = "Patano"
print(a)
print(b)
print(c)
```

#### Unpack a collection
In python, you are allowed to extract values into variables from a list or collection of values.

```python
names = ["Patano", "Michael", "Joyce"]
a, b, c = names
print(names)
print(a)
print(b)
print(c)
```

### Output Variables
In Python, the **print()** funtion allows you to output variables. This is demonstrated by the examples we have been doin so far. Nonetheless, we will do another example now.

```python
a = "Patano"
b = c = d = (10)
print(a)
print(b)
print(c)
print(d)
```

We can also output multiple variables by separating them in the print() function using commas.

```python
a = "My"
b = "name is"
c = "Patano."
print(a, b, c)
```

Python also allows us to use **+** to output multiple variables. Notice the space after **My** and **is** to avoid outputting everything as one.

```python
a = "My "
b = "name is "
c = "Patano."
print(a + b + c)
```

Remember that in python, when using + to output multiple variables made up of values, it performs an addition

```python
a = 10
b = 20
c = 2
print(a + b + c)
```

It is also important to note that when using + to output multiple variables made up of strings and numbers in python, it will result in an error

```python
a = 10
b = "Patano"
print(a + b)
```

It is therefore more appropriate to use comma in outputing multiple variables
