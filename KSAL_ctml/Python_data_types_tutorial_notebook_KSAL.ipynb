```python
# A tutorial on data types and manipulation of data in Python
# In this notebook, we will explore the basic and advanced data types in Python and learn how to manipulate them effectively.

# Welcome to this tutorial on Python data types. In this notebook, we will explore the basic and advanced data types in Python and learn how to manipulate them effectively.

# 1. Basic Data Types

# Integers(int)

# Integers are whole numbers, positive, negative, or zero without decimals.

# Examples of integers
a = 10
b = -5

# Performing arithmetic operations
sum_ab = a + b
difference_ab = a - b

print("Sum: {sum_ab}, Difference: {difference_ab}")

# Floats (float)

# Floats are numbers that have a decimal point.

# Examples of floats
x = 10.5
y = -3.2

# Performing arithmetic operations
product_xy = x * y
quotient_xy = x / y

print(f"Product: {product_xy}, Quotient: {quotient_xy}")

# Strings (str)

# Strings are sequences of characters enclosed in quotes (double or single)

# Examples of strings
greeting = "Hello, fellow SAK members!"
name = 'Koney Shardow Abdul Latif'

# Concatenating strings
message = greeting + " My name is " + name

print(message)

# Booleans (bool)

# Booleans represent the truth values: True and False.

# Examples of booleans
is_sunny = True
is_raining = False

# Logical operations
print(is_sunny and is_raining)  # False
print(is_sunny or is_raining)   # True

# 2. Adanced Data Types

# Lists (list) 

# Lists are ordered collections of items enclosed in square brackets []. Elements can be of different types.

# Examples of lists
numbers = [1, 2, 3, 4, 5]
mixed_list = [1, "hello", 3.5, True] # Mixed data types

# Accessing list elements
first_element = numbers[0]
last_element = mixed_list[-1]

print(f"First: {first_element}, Last: {last_element}")

# Manipulating lists
numbers.append(6)
numbers.remove(2)

print(numbers)

# Tuples (tuple)

# Tuples are ordered, immutable (cannot be changed after creation) collections of items enclosed in parentheses (). 

# Examples of tuples
coordinates = (10.0, 20.0)
colors = ('brown', 'green', 'blue')

# Accessing tuple elements
first_color = colors[0]

print(f"First color: {first_color}")

# Modifying Tuples : Attempting to modify a tuple will result in an error 

# Dictionaries (dict)

# Dictionaries are collections of key-value pairs.

# Examples of dictionaries
person = {"name": "Shardow", "age": 22, "city": "Kuamsi"}

# Accessing dictionary values
name = person["name"]
age = person["age"]

print(f"Name: {name}, Age: {age}")

# Manipulating dictionaries
person["age"] = 22
person["country"] = "Ghana"

print(person)

# Sets

# Sets are unordered collections of unique items enclosed in curly brackets {}. They are useful for removing duplicates  or for checking membership.

# Examples of sets
fruits = {"apple", "banana", "orange"}
numbers = {1, 2, 3, 4, 4, 4}

# Adding and removing elements
fruits.add("pear")
fruits.remove("banana")

print(fruits)


# 3. Data type manipulation

# Converting between types
integer_number = 10
float_number = float(integer_number)  # int to float

string_number = "20"
integer_from_string = int(string_number)  # str to int

print(f"Float: {float_number}, Integer: {integer_from_string}")

# String formatting
name = "Shardow"
age = 22
formatted_string = "My name is {name} and I am {age} years old."

print(formatted_string)


```
