**1. Python Basics** 
*Variables and Data Types*

Variables in Python are used to store data. The type of data determines how the variable behaves. Python has dynamic typing, meaning you don’t need to specify the data type.

```python
#String - a sequence of characters
greeting = "Hello, World!"
print(greeting)  

#Integer - whole numbers
age = 30
print("Age:", age)  

# Float - numbers with decimal points
pi_value = 3.14159
print("Value of Pi:", pi_value) 

# Boolean - True or False
is_raining = False
print("Is it raining?", is_raining)  
```

**Explanation**:

String: Holds textual data.
Integer: Holds whole numbers.
Float: Holds decimal-point numbers.
Boolean: Used for conditions (True/False)

**Another Example**:

You can change the type of a variable in Python easily

```python
variable = 100     
print(variable)    

variable = "A string now!"  
print(variable)    
```

**Basic Arithmetic Operations**

Python can handle simple math operations just like a calculator.
Examples:

```python
# Basic arithmetic
a = 10
b = 4

#Addition
sum_result = a + b
print("Addition:", sum_result)  

# Subtraction
sub_result = a - b
print("Subtraction:", sub_result)  

# Multiplication
mul_result = a * b
print("Multiplication:", mul_result)  

# Division
div_result = a / b
print("Division:", div_result)  

# Modulus (remainder)
mod_result = a % b
print("Modulus:", mod_result)
```

**Another example** 

```python

c = -15
d = 7

# Addition
print(c + d)  

# Multiplication 
print(c * d)  
```

**Conditional Statements (if-else)**

Conditionals allow decisions to be made in code. The if, elif, and else keywords are used to control the flow.

```python
temperature = 22

if temperature > 30:
    print("It's hot outside.")
elif temperature > 20:
    print("The weather is nice.")
else:
    print("It's cold outside.")
```

**Explanation**:

if checks the first condition.
elif checks additional conditions.
else runs if no previous conditions are met.

Another Example:

```python
# Nested if conditions
is_weekend = True
is_raining = True

if is_weekend:
    if is_raining:
        print("Stay inside and watch a movie!")
    else:
        print("Go outside and enjoy the sun!")
else:
    print("Time to work!")
```

**Loops**

Loops are used to repeat a block of code multiple times.
**For** Loop Example:

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

**While** Loop Example:

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

**Functions**

Functions are reusable blocks of code that you can call multiple times.
Examples:

```python
# Simple function to add two numbers
def add_numbers(a, b):
    return a + b

result = add_numbers(10, 5)
print("Sum:", result)  
```

**Explanation**:

def is used to define a function.
return sends back the result.

Another Example:

```python
# Function to check if a number is even or odd
def check_even_odd(number):
    if number % 2 == 0:
        return "Even"
    else:
        return "Odd"


print(check_even_odd(10)) 
print(check_even_odd(7))   
```

**2.0 Object-Oriented Programming (OOP)**
OOP helps to organize code into classes and objects, modeling real-world entities.
A class is a blueprint for creating objects. Each object can have its own properties and behavior.

**Example**

```python
class Car:
    # The __init__ method is called the constructor
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def describe(self):
        return f"This car is a {self.year} {self.make} {self.model}."

# Create an object of Car class
my_car = Car("Toyota", "Corolla", 2020)

# Access object attributes
print(my_car.describe())  
```

**Explanation**:

__init__ is a special method called when the object is created.
self refers to the current instance of the class.

Another Example :

```python
# Another object of the Car class
car2 = Car("Honda", "Civic", 2022)
print(car2.describe())  
```

**Inheritance**

Inheritance allows one class to derive (inherit) properties and methods from another class 

**Example**

```python
# Parent class
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print(f"{self.name} makes a sound.")

# Child class
class Dog(Animal):
    def sound(self):
        print(f"{self.name} barks.")

# Creating a Dog object
dog = Dog("Rex")
dog.sound()  
```

**Explanation**:

Dog class inherits from Animal, but can override methods (e.g., sound).

**Another Example**:

```python
# Another child class
class Cat(Animal):
    def sound(self):
        print(f"{self.name} meows.")

cat = Cat("Whiskers")
cat.sound() 
```

**Encapsulation**

Encapsulation means restricting direct access to some data.
**Example**:

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance

# Create a bank account object
account = BankAccount(1000)
account.deposit(500)
print("Balance:", account.get_balance())  
```

**Polymorphism**

Polymorphism allows the same method name to behave differently for different classes.
**Example**:

```python
class Bird:
    def speak(self):
        print("Bird sings.")

class Dog:
    def speak(self):
        print("Dog barks.")

def make_sound(animal):
    animal.speak()

bird = Bird()
dog = Dog()

make_sound(bird)  
make_sound(dog)  
```
