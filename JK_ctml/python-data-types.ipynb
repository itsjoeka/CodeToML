## LESSON 1
> **Python Variables** 

There is no specific command for assigning python variables. The moment you assign a value to it, it is created

```python
#creating a variable called x and assigning it a value of 5
x = 5

#return the variable to the screen
print(x)
```

```python
#more variables
name = "Pythonista"
age = 26

print(f"I will be a {name} at the age of {age}")
```

```python
#casting allows us to specify the data type for our variable
y = str(20)
print(type(y))
```

### variable nomencleture
- variable name must start with an **alphabet** or the **underscore character** and cannot start with a number
- variable names are case sensitive (X and x are different variable names)
- variable names can only contain **alphanumeric** characters and **underscores**
- variable names can not be any of the **python** keywords

```python
#assigning multiples variables with values in a single line of code
_joe, edem, ben_1, eben2 = 23, 26, 13, 16

print(_joe + edem + ben_1 + eben2)
```

```python
#assigning one value to multiple variables

a = b = c = True
print(a)
print(b)
print(c)
```

```python
#assigning the values of a collection to multiple variables
regions = ["ashanti", "volta", "eastern", "northen"]
d, e, f, g = regions
print(d)
print(e)
print(f)
print(g)
```

### STRINGS
Python strings are enclosed in either double or single quotation marks

```python
#string and returning string literal
print("Jonathan") #is a string
surname = "Kalami"
print(surname)
```

```python
#you can type a multiline string using three double quotation marks
"""
Dear Sir,
I am writing to express my interest in the biomedical informatics PhD program 
that was recently advertised on your laboratory website.
Thank you

yours faithfully.
"""
```

```python
#printing same in multiline. escaping with \n
print("Dear Sir,\nI am writing to express my interest in the biomedical informatics PhD program\nthat was recently advertised on your laboratory website.\nThank you.\n yours faithfully.")
```

```python
#typing quotes in the quotes. Use double quotes to escape single quotes and vice versa
print("Hello in Deutsch is 'Hallo'")
print('Hallo in English is "Hello"')
```

```python
numstr = '2, 3, 4, 5' #any value enclosed in quotations are strings no matter their type
print(type(numstr))
```
