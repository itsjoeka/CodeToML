There is no specific command for assigning python variables. The moment you assign a value to it, it is created

### variable nomencleture
- variable name must start with an **alphabet** or the **underscore character** and cannot start with a number
- variable names are case sensitive (X and x are different variable names)
- variable names can only contain **alphanumeric** characters and **underscores**
- variable names can not be any of the **python** keywords

```python
#assigning multiples variables with values in a single line of code
_joe, edem, ben_1, eben2 = 23, 26, 13, 16

print(_joe + edem + ben_1 + eben2)
```

```python
#assigning one value to multiple variables

a = b = c = True
print(a)
print(b)
print(c)
```

```python
#assigning the values of a collection to multiple variables
regions = ["ashanti", "volta", "eastern", "northen"]
d, e, f, g = regions
print(d)
print(e)
print(f)
print(g)
```

## LESSON 2
> **Python Data Types:** 

There are a number of built-in python **data types**
- Strings or str
- Integers or int
- Floats
- Complex
- List
- Tuple
- Set
- Frozenset 
- Dictionary
- Boolean
- Bytes
- Bytearray
- Memoryview

```python
#STRINGS
"""
a string is a python data type that can be created by enclosing the characters in a double or singe quotation 
mark. use the double quotation to escape intra single quotation marks and vice versa
"""
a_string = "Hello world"
print(type(a_string))
print()

#INTEGER
"""
integers are python data types that represent natural numeric figures
"""
num = 5
print(type(num))
print()

#FLOAT
"""
floats are python data types that represent the collection of mathematical decimal figures
"""
flp = 4.7
print(type(flp))
print()

#BOOLEAN
"""
boolean values are either True or False. An expression in python could either be True or False
"""
boole1 = True
boole2 = False
print(f"{boole1} or {boole2}")
print(bool(1))
print(bool(0))
print(type(boole1))
print()

#it is possible to assign the data type of a given value you wish to create
assign = str(2)
print(assign)
print(type(assign))

figure = 70
print(type(figure))
print(float(figure))
print(str(figure))
print(bool(figure))
```
