# **Object Oriented Programming (OOP)**

OOP is a **programming paradigm** that provides a means of structuring programs so that `properties` and `behaviours` are bundled together.

CONCEPT OF `OOP` => `PROPERTIES` + `BEHAVIOURS` = `STRUCTURED PROGRAM`

OOP allows one to model real world things like cars, relationships e.t.c through a structural programming.

The four Tenets of OOP are:
1. **Encapsulation**: allows you to bundle data (attributes) and behaviours (methods) withing a class to create a cohesive unit.
2. **Inheritance**: enables the creation of hierarchical relationships between classes, allowing a subclass to inherit attributes and methods from a parent class.
3. **Abstraction**: shifts focus from implementation details to essential functionality. Focuses on what an object does rather than how it does it.
4. **Polymorphism**: allows you to treat objects of different types as instances of the same base type, as long as they implement a common interface or behaviour.

## How do Define a Class in Python

The keyword `class` is used to define a class followed by the name of the class and a colon.

### **\_\_init\_\_** & **attributes**

This is used to declare which attributes (data) each instance of the class should have.

We want to spice the `Dog` class up a bit by defining some properties that all Dog objects should have (e.g name, age, coat color, breed e.t.c.).
- The `__init__()` method is used to define all the properties you want the object to have.
- The `__init__()` method sets the initial state of every new object created by assigning the values of the object's properties
- So it initializes each new instance
- First parameter of the `__init__()` method is always the `self` parameter so that python can define new atttributes on the object.

````python
class Dog:
    def __init__(self, name, age):  
        self.name = name            # receives data for the name of an instance
        self.age = age              # receives data for the age of an instance
````

* **self.name = name** creates an attribute called `name` and assigns the value of the `name` parameter to it
* **self.age = age** creates an attribute called `age` and assigns the value of the `age` parameter to it.

**Note**: the .name after `self` is arbitrary so is the .age. These are the defined parameters. The items after the equal sign however has to match the words in the attribute bracket of the defined method.

Types of Attributes:
- **Instance attribute**: created in the `__init__()` method. This is specific to a particular instance of the class. Different instances will have different values.
- **Class attribute**: attributes that have the same value for all class instances. Defined outside of the `__init__()` method. 

````python
class Dog:
    species = "Canis familiaris"    # class attribute

    def __init__(self, name, age):  # instance attribute
        self.name = name
        self.age = age
````

### **Instance Methods**

**Instance methos** are functions that you define inside a class and can only call on an instance of that class. Just like the `__init__()` method, an instance method always take `self` as its first parameter.

````python
class Dog:
    # class attribute
    species = "Canis familiaris"

    # Instance method (for attributes)
    def __init__(self, name, age):
        # Instance attributes
        self.name = name
        self.age = age

    # Instance method
    def description(self):
        return f"{self.name} is {self.age} years old"

    # Another instance method
    def speak(self, sound):
        return f"{self.name} says {sound}"
````

This dog class has two instance methods
* **.description()**: returns a string displaying the name and age of the dog
* **.speak()**: has one parameter called `sound` and returns a string containing the dog's name and the sound that the dog makes

````python
# instantiate (or create an instance of the class)
sky = Dog("Sky", 4)

# return attributes
print("All details about this dog:\n", f"Species: {sky.species}")
print(f" Name: {sky.name}")
print(f" Age: {sky.age}")
print()

# perform methods
print(sky.description())        # methods takes no argument
print(sky.speak("woof woof"))   # method takes one argument
````

You can change the output `print` returns when called on class instances by defining the `__str__()` method

````python
# define a class
class Dog:
    species = "Canis familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"

    def speak(self, sound):
        return f"{self.name} says {self.sound}"


# create an instance of the class
sky = Dog("Sky", 4)

# perform a print on the instance created
print(sky)
````

Methods like `__init__()` and `__str__()` are called **dunder methods** because they begin and end in double underscores. There are plenty of them that you can use to customize a class in python.

## Interitance from Another Class

**Inheritance** is the process by which one class takes on the attributes and methods of another. Newly formed classes are called **child classes**, and the classes that you derive child  classes from are called **parent clases**

You inherit from a parent class by creating a new class and putting the name of the parent class into parentheses:

````python
# define parent class
class Parent:
    hair_color = "brown"

# create an inheritance from parent class
class Child(Parent):
    pass
````

````python
print(Parent.hair_color)
print(Child.hair_color)
````

````python
class Shape:
    def __init__(self, color, is_filled):
        self.color = color
        self.is_filled = is_filled

class Circle(Shape):
    def __init__(self, color, is_filled, radius):
        super().__init__(color, is_filled)
        self.radius = radius

class Square(Shape):
    def __init__(self, color, is_filled, width):
        super().__init__(color, is_filled)
        self.width = width

class Triangle(Shape):
    def __init__(self, color, is_filled, height, width):
        super().__init__(color, is_filled)
        self.height = height
        self.width = width


t = Triangle(color="red", is_filled=True, height=3, width=2)
print(t.color)
print(t.is_filled)
````
