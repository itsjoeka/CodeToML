## Python Basics 

### Variables 
A __*variable*__ is a placeholder for information you want Python to recall later in the coding process when you need to complete an action. Technically, the variable acts as an address for where the data is stored in memory.

```python
# Examples of variables 
y = 6
a = "shardows"
# in this case the variables y and a were created and assigned different values of different types
```

### Data Types

#### Integers (int)
__*Integers*__ are whole numbers, positive, negative, or zero without decimals.


```python
# Examples of integers
a = 7
b = -9
```

```python
# Performing arithmetic operations
product_ab = a * b
quotient_ab = a / b

print(f"Product: {product_ab}, Quotient: {quotient_ab}")
```

#### Strings (str)
__*Stings*__ are sequences of characters enclosed in quotes (double or single)

```python
# Examples of strings
Welcome_message = "Hello, fellow CodeToML colleagues!"
name = 'Koney Shardow Abdul Latif'
```

### Operators
__*Operators*__ are special symbols in Python that perform specific operations on variables and values. Python supports various types of operators, which we’ll explore below.

#### Arithmetic Operators
These operators are used for mathematical operations.

-  Addition (+)
-  Subtraction (-)
-  Multiplication (*)
-  Division (/)
-  Floor Division (//)
-  Modulus or Remainder (%)
-  Exponentiation (**)

```python
# Examples of mathematical operations using arithmetic operators
x = 10
y = 3
print(x + y)  # Output: 13
print(x % y)  # Output: 1
print(x ** y)  # Output: 1000 (10 raised to the power of 3)
```

#### Comparison (Relational) Operators
Comparison operators compare two values and return *True* or *False*.

-  Equal to (==)
-  Not equal to (!=)
-  Greater than (>)
-  Less than (<)
-  Greater than or equal (>=)
-  Less than or equal (<=)

```python
# Examples
a = 10
b = 5
print(a > b)  # Output: True
print(a == b)  # Output: False
```

#### Logical Operators
Logical operators are used to combine conditional statements.

- and (Returns *True* if both statements are True)
- or (Returns *True* if one of the statements is True)
- not (Reverses the result: *True* becomes *False*)

```python
# Examples 
a = 7
print(a > 5 and a < 10)  # Output: True
print(not(a == 7))  # Output: False
```

#### Assignment Operators
Assignment operators are used to assign values to variables.

- = (assigns a value to a variable)
- +=
- -=
- *=
- /=
- **=

```python
# Examples 
x = 5
x += 2  # Same as x = x + 2
print(x)  # Output: 7
```

### Control Flow
__*Control flow*__ allows you to dictate how the code should execute based on certain conditions or repeat certain actions.

#### Conditional Statements
Conditional statements allow you to make decisions in your code. The most common conditional statement in Python is the *if* statement.

-  *if* condition (Block of code to execute if condition is *True*)
-  *elif* another_condition (Block of code if this condition is *True*)
-  *else* Block of code if none of the above conditions are True

```python
# Examples
age = 25

if age < 18:
    print("You are a minor.")
elif age == 18:
    print("You just turned 18!")
else:
    print("You are an adult.")
```

*Comparison with Logical Operators*
<br>You can combine conditions using logical operators (and, or, not).

```python
# Examples 
age = 26
if age > 18 and age < 27:
    print("You are a young adult.")
```

#### Loops
Loops allow you to execute a block of code repeatedly as long as a condition holds true.

**For Loop**
<br> A *for* loop is used to iterate over a sequence (like a list, tuple, or string).
<br> *for* item *in* iterable: (Block of code to repeat for each item)

```python
# Examples 
fruits = ["apple", "banana", "orange"]
for fruit in fruits:
    print(fruit)
```

```python
# More Examples using the range() function
for u in range(5):
    print(u)
```

**While Loop**
<br> A *while* loop continues executing as long as a condition remains true.
<br> *while* condition: (Block of code to repeat while the condition is True)

```python
# Examples 
count = 0
while count < 5:
    print(count)
    count += 1  # Increment the counter to avoid an infinite loop
```

**Break** and **Continue**
<br> You can control the flow of a loop using *break* and *continue*.
-  *break* : (Exits the loop completely).
-  *continue* : (Skips the current iteration and continues with the next one).

```python
# Examples 
for i in range(10):
    if i == 5:
        break  # Exit the loop when i equals 5
    print(i)
```

## Python Functions
A __*function*__ is a block of code designed to perform a specific task. Functions help to make the code modular, reusable, and organized.

**Why Use Functions?**
-  *Reusability* : Functions allow you to reuse code instead of writing it multiple times.
-  *Modularity* : Code becomes easier to read and maintain when divided into smaller, manageable functions.
-  *Maintainability* : Debugging becomes simpler because each function performs a well-defined task.


**Defining a Function**
In Python, functions are defined using the *def* keyword, followed by the function name and parentheses.

def function_name(parameters):
    """Optional docstring: Explain what the function does)

```python
# Example 
def welcome():
    """This function prints a welcome message."""
    print("Hello, welcome to CodeToML!")

welcome ()  # Calling the function
```

**Parameters and Arguments**
<br>__*Parameters*__ are variables defined in the function's parentheses that allow you to pass information into the function.
<br>__*Arguments*__ are the actual values passed to the function when it is called.

```python
# Example
def greet_user(name):
    """This function greets the user by name."""
    print(f"Hello, {name}!")

greet_user("Koney")  # Passing an argument
```

**Return Statement**
<br>A function can return a value using the *return* statement. This allows you to store the result of the function and use it elsewhere in the program.

```python
# Example 
def add_numbers(a, b):
    """This function adds two numbers and returns the result."""
    return a + b

result = add_numbers(5, 3)  # Storing the returned value
print(result)
```

### Types of Functions
#### Positional Arguments
In this case, arguments are matched to parameters based on their position.

```python
# Example 
def subtract(a, b):
    return a - b

print(subtract(10, 5))  # Output: 5
```

#### Keyword Arguments
You can specify which arguments to pass by name, allowing you to call functions in a more readable way.

```python
# Example 
def describe_person(name, age):
    print(f"{name} is {age} years old.")

describe_person(age=25, name="Koney")  # Keyword arguments
```

#### Default Arguments
You can assign default values to parameters. If the caller does not provide a value, the default is used.

```python
# Example 
def greet_user(name="Guest"):
    print(f"Hello, {name}!")

greet_user()  # Uses the default value "Guest"
greet_user("Koney")  # Uses the provided argument
```

#### Variable-Length Arguments
**args* for positional variable-length arguments and ***kwargs* for keyword variable-length arguments.

```python
# Example of *args
def add_numbers(*args):
    """This function adds all numbers passed as arguments."""
    return sum(args)

print(add_numbers(1, 2, 3))  # Output: 6
print(add_numbers(10, 20, 30, 40))  # Output: 100
```

```python
# Example of **kwargs
def describe(**kwargs):
    """This function describes a person using keyword arguments."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

describe(name="Koney", age=25, profession="Research Assistant")
```

### Scope of Variables
#### Local Scope
Variables declared inside a function have local scope, meaning they are only accessible within that function.

```python
# Example
def my_function():
    x = 10  # Local variable
    print(x)

my_function()  # Works fine
# print(x)  # Raises an error because 'x' is not accessible outside the function
```

#### Global Scope
Variables declared outside any function have global scope and can be accessed by any function.

```python
# Example
x = 20  # Global variable

def my_function():
    print(x)  # Accessing the global variable

my_function()
print(x)  # Works fine
```

### Lambda Functions
A lambda function is a small anonymous function that can take any number of arguments but has only one expression.
<br> *lambda* arguments: expression

```python
# Example
add = lambda x, y: x + y
print(add(5, 3))  # Output: 8
# Lambda functions are often used for short, simple operations like in sorting or filtering data.
```

## Introduction to Object-Oriented Programming (OOP)
**Object-Oriented Programming (OOP)** is a programming paradigm that organizes software design around data, or objects, rather than functions and logic. OOP makes code reusable and easier to manage, especially for large-scale applications. It also closely resembles how real-world entities work, which makes it intuitive for complex systems.

<br> **What is OOP?** 
<br> In OOP, the key idea is to represent real-world objects and concepts using **classes** and **objects**. This approach allows programmers to group related data (attributes) and behaviors (methods) into a single unit.

<br> **Key Concepts of OOP**
- __*Class*__: A blueprint or template for creating objects (instances).
- __*Object*__: An instance of a class that contains attributes and methods.
- __*Inheritance*__: A mechanism to create a new class from an existing class.
- __*Polymorphism*__: The ability to process objects differently depending on their class.
- __*Encapsulation*__: Bundling data and methods into a single unit (class) and restricting access to some of the object's components.
- __*Abstraction*__: Hiding the implementation details and showing only the functionality to the user.

### Creating Classes and Objects
A **class** defines the structure and behavior of objects. An object is an instance of a class, which means it represents one specific entity created from that blueprint.

#### Defining a Class:
To create a class in Python, use the *class* keyword, followed by the class name. Class names are usually written in PascalCase.

```python
# Example 
class Car:
    """A simple class representing a car."""
    
    # Attributes (variables)
    def __init__(self, make, model, year):
        """Initialize the car with make, model, and year attributes."""
        self.make = make
        self.model = model
        self.year = year
    
    # Method (function)
    def start_engine(self):
        """Simulate starting the car engine."""
        print(f"My {self.year} {self.make} {self.model}'s engine is now running!")

# Creating an object (instance) of the Car class
my_car = Car("Benz", "C300", 2025)

# Accessing the object's attributes and methods
print(my_car.make)  # Output: Toyota
my_car.start_engine()  # Output: The 2020 Toyota Camry's engine is now running!
```

__*Explanation*__
-  The *Car* class has three attributes: *make*, *model*, and *year*.
-  The *start_engine* method is defined to simulate the action of starting the engine.
-  The *my_car* object is an instance of the *Car* class, and we can access its attributes and methods.

### Constructors and Methods
#### Constructors (___*init*___ method):
A __*constructor*__ is a special method called when an object is created. In Python, the constructor is defined by the ___*init*___() method. It is used to initialize the object’s attributes.

```python
# Example 
class Person:
    def __init__(self, name, age):
        """Constructor to initialize name and age."""
        self.name = name
        self.age = age

    def greet(self):
        """Method to greet the person."""
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Creating an object of the Person class
person1 = Person("Koney", 25)
person1.greet()  # Output: Hello, my name is Koney and I am 25 years old.
```

#### Methods
Methods are functions defined inside a class that operate on objects of that class. Methods allow objects to exhibit behavior.

### Inheritance and Polymorphism
#### Inheritance
__*Inheritance*__ allows a new class (child class) to inherit attributes and methods from an existing class (parent class). This promotes code reusability.

<br> *class* ParentClass:
    <br> (Parent class attributes and methods)

*class* ChildClass(ParentClass):
    <br>(Child class inherits from ParentClass)
   <br>(Child class can have additional attributes and methods)

```python
# Example 
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} makes a sound.")

class Dog(Animal):  # Dog class inherits from Animal
    def speak(self):
        print(f"{self.name} barks.")

class Cat(Animal):  # Cat class inherits from Animal
    def speak(self):
        print(f"{self.name} meows.")

dog = Dog("Buddy")
cat = Cat("Whiskers")

dog.speak()  # Output: Buddy barks.
cat.speak()  # Output: Whiskers meows.
```

__*Explanation*__
-  The *Animal* class is the parent class, and both *Dog* and *Cat* classes inherit from it.
-  The *Dog* and *Cat* classes override the *speak*() method, demonstrating polymorphism.

#### Polymorphism:
__*Polymorphism*__ allows methods in different classes to have the same name but behave differently based on the object's class.

```python
# Example 
def animal_sound(animal):
    animal.speak()

dog = Dog("Buddy")
cat = Cat("Whiskers")

animal_sound(dog)  # Output: Buddy barks.
animal_sound(cat)  # Output: Whiskers meows.
```

__*Explanation*__
Here, the *animal_sound*() function works with any object that has a *speak*() method, demonstrating polymorphism.
